```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1.2, 1000)

def f(x):
    return 2*x**2 / (1 + x**3)

def plot_graph():
    plt.plot(x, f(x))
    plt.plot(x, x, color='black', linestyle='dashed', linewidth=0.5)
    
plot_graph()
```

```python
def cobweb(start, N=100):
    plot_graph()
    tx, ty = start, f(start)
    plt.scatter(tx, ty, color='black', s=20, zorder=10)

    for _ in range(N):
        fy = f(ty)
        plt.plot([tx, ty], [ty, ty], color='red')
        plt.plot([ty, ty], [ty, fy], color='red')
        tx, ty = ty, fy
```

```python
# Convergence to x=0
cobweb(0.3)
```

```python
# Convergence to x=1
cobweb(0.8)
```

```python
# Let's try going from the right too
cobweb(1.2)
```

```python
# Even if we start really close to the middle fixed point, we don't converge
# Here we start at the black dot, but end up at x=0
cobweb(0.618)
```
