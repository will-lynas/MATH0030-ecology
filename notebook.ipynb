```python
import numpy as np
import matplotlib.pyplot as plt

def plot_phase_portrait(a, b, c, d, e, f):

    def F(x, y):
        return x * (a - b*x + c*y)

    def G(x, y):
        return y * (d + e*x - f*y)
        
    x_max = 2*a/b
    y_max = 2*d/f

    # fixed points 
    plt.scatter(0, 0, color='red')
    plt.scatter(a/b, 0, color='purple')
    plt.scatter(0, d/f, color='purple')
    if c*e < b*f:
        s = 1/(b*f-c*e)
        plt.scatter(s*(a*f+c*d), s*(b*d+e*a), color='purple')
        x_max = 2*s*(a*f+c*d)
        y_max = 2*s*(b*d+e*a)

    x = np.linspace(0, x_max, 100)
    y = np.linspace(0, y_max, 100)
    X, Y = np.meshgrid(x, y)
    U = F(X, Y)
    V = G(X, Y)
    plt.streamplot(X, Y, U, V, density=2, linewidth=1)

    plt.xlim(0, x_max)
    plt.ylim(0, y_max)


    # nullclines
    plt.plot(x, (b*x - a)/c, color='purple')
    plt.plot(x, (d + e*x)/f, color='purple')

    plt.show()

```

```python
plot_phase_portrait(a=1, b=1, c=1, d=1, e=1, f=1)
```

```python
plot_phase_portrait(a=1, b=1, c=0.2, d=1, e=0.2, f=1)
```

```python
plot_phase_portrait(a=1, b=1, c=0.2, d=0, e=0.2, f=1)
```
